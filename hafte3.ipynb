```python
import nltk
nltk.download('punkt')
```

```python
paragraph = """History, the study of past events, civilizations, and cultures, is a fundamental discipline that enriches our understanding of human society and its development. The importance of history lies not only in chronicling what has occurred but also in shaping our present and guiding our future. It offers invaluable lessons, fosters a sense of identity, and provides perspective on contemporary issues.

The importance of history begins with its role in understanding human development. By examining the origins and progress of different societies, we gain insights into the factors that have shaped the world. From the rise and fall of empires to the impact of revolutions and technological advancements, history provides a comprehensive narrative of human evolution. For instance, studying the Industrial Revolution helps us understand the transition from agrarian societies to industrialized nations and its effects on social structures, economy, and culture.

History also serves as a repository of collective memory. It preserves the experiences, achievements, and failures of past generations, allowing us to learn from them. This collective memory helps prevent the repetition of past mistakes. The horrors of wars, genocides, and oppressive regimes are documented in history, serving as stark reminders of the consequences of hatred, intolerance, and unchecked power. By reflecting on these events, societies can strive to avoid similar atrocities in the future.

Moreover, history fosters a sense of identity and continuity. It connects individuals to their cultural heritage, instilling a sense of belonging and pride. For nations and communities, history is a source of shared narratives that reinforce collective identity. National holidays, commemorative monuments, and historical literature all play a role in keeping historical consciousness alive, helping people understand their place within the broader tapestry of human history. This connection to the past is essential for preserving cultural diversity and fostering mutual respect among different communities.

Understanding history also provides context for contemporary issues. Many current global challenges, such as political conflicts, economic disparities, and social movements, have deep historical roots. By analyzing the historical context of these issues, we can gain a better understanding of their complexities and develop more informed solutions. For instance, the ongoing conflict in the Middle East cannot be fully comprehended without knowledge of its historical background, including colonialism, religious tensions, and geopolitical strategies.

In addition, history enhances critical thinking and analytical skills. Historians engage in rigorous analysis of sources, assessing their reliability and bias. This process cultivates the ability to think critically about information, an essential skill in the age of information overload and misinformation. By learning to scrutinize historical evidence and arguments, individuals develop a more discerning approach to contemporary information and media.

Furthermore, history inspires innovation and progress by providing a foundation of knowledge and experience. Many scientific and technological advancements build upon historical discoveries and inventions. For example, modern medical practices are rooted in historical research and experimentation. By studying the achievements of the past, innovators can build on previous knowledge, avoid past errors, and explore new possibilities.

In conclusion, history is a vital discipline that offers a window into the human experience, providing lessons, identity, context, and critical thinking skills. Its study is essential for understanding our past, navigating our present, and shaping our future. By valuing and learning from history, we can cultivate a more informed, empathetic, and forward-looking society."""
```

```python
# Tokenizing sentences
sentences = nltk.sent_tokenize(paragraph)
sentences
```

```python
#Tokenizing words
words = nltk.word_tokenize(paragraph)
words
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stemmer = PorterStemmer()

#stemming
for i in range (len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [stemmer.stem(word) for word in words if word not in set (stopwords.words('english'))]
  sentences[i] = ' '.join(words)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)

#lemmatization
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [lemmatizer.lemmatize(word) for word in words if word not in set (stopwords.words('english'))]
  sentences[i]= ' '.join(words)
```

```python
#cleaning the texts
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

ps = PorterStemmer()
wordnet = WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]',' ',sentences[i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word)for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)

#creating the bag of words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
x = cv.fit_transform(corpus).toarray()
```
